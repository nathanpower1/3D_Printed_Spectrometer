```python
def slit_func(z,wave_l,L):
    return (L*wave_l) / (z)
L = 113.5
w = 632E-9
z = np.array([0,3.5,1.8,1.2,0.9,0.7])
a = np.array([0,2E-5,4E-5,6E-5,8E-5,10E-5])
s =slit_func(z,w,L)
x = np.linspace(0,5,6)
plt.scatter(x,s, color="r")
plt.plot(x,a)
plt.ticklabel_format(axis='both', style='sci', scilimits=(0,0))
plt.grid(True)
plt.title("Calibration Plot for Variable Slit")
plt.xlabel("Order Number")#
plt.ylabel("Distance from Centre")
plt.legend(["Calculated Values", "Expected Values"])
```
